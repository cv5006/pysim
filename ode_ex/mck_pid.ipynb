```python
import numpy as np
import pandas as pd
from math import pi, sin
from scipy.integrate import solve_ivp
import plotly.express as px

from mck_model import *

x0 = [0, 0]
t_start, t_end = 0, 10
t = np.linspace(t_start, t_end, 1000)

m = 2
c = 1
k = 1
mck = MCKmodel(m, c, k, 0)

mck.setU(lambda t, x: sin(2*pi*t))

sol = solve_ivp(mck.model, [t_start, t_end], x0, method='RK45', t_eval=t)

df = pd.DataFrame(data={'Time':t, 'Position':sol.y[0,:], 'Velocity':sol.y[1,:]})
px.line(df, x='Time', y='Position')
```
