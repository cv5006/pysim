```python
import numpy as np
import pandas as pd
from scipy.integrate import odeint
import plotly.express as px

# free fall equation
def FreeFall(x, t, m):    
    b = 0.003
    g = 9.81
    dxdt = x[1], b/m*x[1]*x[1] -g
    return dxdt


p0 = 100
v0 = 10
x0 = [p0, v0]

t = np.linspace(0, 20, 1000)

m = 2
x = odeint(FreeFall, x0, t, args=(m,))

df = pd.DataFrame(data={'Time':t, 'Position':x[:, 0], 'Velocity':x[:, 1]})

px.line(df, x='Time', y='Velocity')
```
